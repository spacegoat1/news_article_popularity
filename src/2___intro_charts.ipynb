```python
from project_paths import *
import pandas as pd
import matplotlib
from matplotlib import pylab as plt
```

```python
data = [{'year': 2011, 'share_revenue_digital_advertising': 0.17},
    {'year': 2012, 'share_revenue_digital_advertising': 0.19},
    {'year': 2013, 'share_revenue_digital_advertising': 0.20},
    {'year': 2014, 'share_revenue_digital_advertising': 0.21},
    {'year': 2015, 'share_revenue_digital_advertising': 0.25},
    {'year': 2016, 'share_revenue_digital_advertising': 0.29},
    {'year': 2017, 'share_revenue_digital_advertising': 0.31},
    {'year': 2018, 'share_revenue_digital_advertising': 0.35},
    {'year': 2019, 'share_revenue_digital_advertising': 0.35},
    {'year': 2020, 'share_revenue_digital_advertising': 0.39}]
```

```python
df = pd.DataFrame(data)
df["share_revenue_digital_advertising"] = df["share_revenue_digital_advertising"]*100
df = df.set_index("year")
```

```python
df.plot.line(figsize=(9,5), legend=False)
plt.xlabel("Year")
plt.ylabel("Percentage")
# plt.legend(hide=True)
plt.title("Share of advertising revenue from digital advertising")
plt.rcParams["figure.dpi"] = 300
plt.savefig(figures_dir + '0_advert_revenue.png')
```
